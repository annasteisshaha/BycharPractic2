```python
from my_func import*
import os
```

```python
list_func = ['0 : Узнать директорию',
            '1 : Создать директорию',
            '2 : Удалить директорию',
            '3 : Изменить директорию',
            '4 : Создать файл',
            '5 : Записать данные в файл',
            '6 : Считать данные из файла',
            '7 : Удалить файл',
            '8 : Копировать файл',
            '9 : Переместить файл',
            '10 : Переименовать файл',
             '11 : Завершить работу файлового менеджера']
def get_dir():
    direct = os.getcwd()
    print('Вы находитесь в папке' , direct.split('\\')[-1])
```

```python
for _ in list_func:
        print(_)
while True:
    print('Выберете действие:')
    act = input('Введите номер действия ->> ')
    if act == '0':
        get_dir()
    elif act == '1':
        name = input('Введите имя папки ->> ')
        if name == 'отмена':
            continue
        else:
            create_folder(name)
    elif act == '2':
        name = input('Введите имя папки для удаления ->> ')
        if name == 'отмена':
            continue
        else:
            delete_folder(name)
    elif act == '3':
        name = input('Введите имя папки, в которую хотите перейти ->> ')
        if name == 'отмена':
            continue
        else:
            change_directory(name)
    elif act == '4':
        name = input('Введите имя файла ->> ')
        if name == 'отмена':
            continue
        else:
            create_file(name)
    elif act == '5':
        name = input('Введите имя файла ->> ')
        if name == 'отмена':
            continue
        else:
            text = input('Введите текст для записи ->> ')
            add_text_to_file(name, text)
    elif act == '6':
        name = input('Введите имя файла ->> ')
        if name == 'отмена':
            continue
        else:
            show_text_file(name)
    elif act == '7':
        name = input('Введите имя файла ->> ')
        if name == 'отмена':
            continue
        else:
            delete_file(name)
    elif act == '8':
        name = input('Введите имя файла для копии ->> ')
        if name == 'отмена':
            continue
        else:
            name_dir = input('Введите имя папки ->> ')
            copy_file(name, name_dir)
    elif act == '9':
        name = input('Введите имя файла для перемещения ->> ')
        if name == 'отмена':
            continue
        else:
            name_dir = input('Введите имя папки ->> ')
            move_file(name, name_dir)
    elif act == '10':
        name = input('Введите имя файла для смены имени ->> ')
        if name == 'отмена':
            continue
        else:
            new_name = input('Введите новое имя ->> ')
            rename_file(name, new_name)
    elif act == '11':
        print('Работа файлового мнеджера завершена!')
        break
    else:
        print('Неккоректный ввод!')
        continue
```
